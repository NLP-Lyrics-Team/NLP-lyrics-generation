Read the Genius fragments to retrieve their artists

```python
import glob
import pandas as pd

# Set pandas option
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)


# Get input csv filenames of genius fragments
filenames = [i for i in glob.glob(fragments_path)]
print("fragments filenames:", filenames)

# Read CSV files and get columns of interest
data = pd.concat([pd.read_csv(f) for f in filenames], ignore_index=True)
data = data[['artist', 'genre']]

# Keep one entry per artist
data.drop_duplicates(subset='artist', inplace=True)

# Remove 'Other' genre
data = data[data.genre != "Other"]
#data = data[data.genre.str.lower() == "not available"]
#data = data[data.genre == "Not available"]

# Keep only artists
original_data = data['artist']

original_artists = set()
original_data.apply(lambda x: original_artists.add(x))

print("Artists:")
print(original_data)
```

Read the Genius dataset to compare artists

```python
# Read CSV file and get columns of interest
genius_data = pd.read_csv(genius_path)
genius_data = genius_data['artist']

# Keep one entry per artist
genius_data.drop_duplicates(inplace=True)

genius_artists = set()
genius_data.apply(lambda x: genius_artists.add(x))

print("Genius artists:")
print(genius_data)
```

Check via sets

```python
# FUNZIONA SOLO PER CONFRONTARE IL DATASET MERGE GENIUS CON I VARI FRAMMENTI GENIUS PERCHÈ I NOMI DEGLI ARTISTI NON SONO SCRITTI IN MODO UGUALE A METROLYRICS
miss = original_artists.difference(genius_artists)

print("Missing artists in genius csv:")
for artist in miss:
  print(artist)
print("\nSize: ", len(miss))
```

Check with pandas

```python
# FUNZIONA SOLO PER CONFRONTARE IL DATASET MERGE GENIUS CON I VARI FRAMMENTI GENIUS PERCHÈ I NOMI DEGLI ARTISTI NON SONO SCRITTI IN MODO UGUALE A METROLYRICS
merged_data = pd.merge(original_data, genius_data, how='outer', indicator=True)
data_not_in_genius_csv = merged_data[merged_data['_merge'] == 'left_only']

print("Missing artists in genius csv:")
print(data_not_in_genius_csv)
print("\nSize: ", len(data_not_in_genius_csv.index))
```

Read the preprocessed dataset and display unique genres in it
